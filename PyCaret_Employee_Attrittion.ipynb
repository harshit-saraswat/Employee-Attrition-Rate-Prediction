```
# Neccessary Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pycaret.regression import *

pd.set_option('display.max_columns', 50)
```

```
df_train=pd.read_csv("/content/Train.csv")
df_train.drop('Employee_ID',axis=1,inplace=True)
df_train.head()
```

```
df_test=pd.read_csv("/content/Test.csv")
df_test.drop('Employee_ID',axis=1,inplace=True)
df_test.head()
```

```
trainData = setup(df_train, target = 'Attrition_rate', categorical_imputation='mode')
```

```
compare_models(sort='RMSE')
```

```
# Lasso Regression
lassoReg=create_model('lasso')
```

```
# ElasticNet Regression
enetReg=create_model('en')
```

```
# Lasso Least Angle Regression
larLassoReg=create_model('llar')
```

```
# Bayesian Ridge Regression
bRidgeReg=create_model('br')
```

```
# Huber Regression
hubReg=create_model('huber')
```

```
# Tune Lasso model
tunedLassoBag=tune_model('lasso',ensemble=True,method='Bagging')
```

```
# Tune Lasso model
tunedLassoBoost=tune_model('lasso',ensemble=True,method='Boosting')
```

```
# Tune Enet Model
tunedEnetBag=tune_model('en',ensemble=True,method='Bagging')
```

```
# Tune Enet Model
tunedEnetBoost=tune_model('en',ensemble=True,method='Boosting')
```

```
# Tune LLAR Model
tunedLLARBag=tune_model('llar',ensemble=True,method='Bagging')
```

```
# Tune LLAR Model
tunedLLARBoost=tune_model('llar',ensemble=True,method='Boosting')
```

```
# Tune BRidge Model
tunedBRidgeBag=tune_model('br',ensemble=True,method='Bagging')
```

```
# Tune BRidge Model
tunedBRidgeBoost=tune_model('br',ensemble=True,method='Boosting')
```

```
# Tune Huber Model
tunedHuberBag=tune_model('huber',ensemble=True,method='Bagging')
```

```
# Blend Models
blendModel=blend_models(estimator_list=[lassoReg,enetReg,larLassoReg,bRidgeReg,hubReg])
```

```
# Tune Lasso model
tunedLasso=tune_model('lasso',n_iter=100,optimize='mse')
```

```
# Tune BRidge model
tunedbRidge=tune_model('br',n_iter=100,optimize='mse')
```
